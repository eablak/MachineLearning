```python
def preprocess(df):
    df = df.copy()
    
    def normalize_name(x):
        return " ".join([v.strip(",()[].\"'") for v in x.split(" ")])
    
    def ticket_number(x):
        return x.split(" ")[-1]
    
    def ticket_item(x):
        items = x.split(" ")
        if len(items) == 1:
            return "NONE"
        return "_".join(items[0:-1])
    
    df["Name"] = df["Name"].apply(normalize_name)
    df["Ticket_number"] = df["Ticket"].apply(ticket_number)
    df["Ticket_item"] = df["Ticket"].apply(ticket_item)
    return df

preprocessed_train_df = preprocess(train)
preprocessed_serving_df = preprocess(test)

preprocessed_train_df.head()
```

```python
train.isnull().sum().sort_values(ascending=False).to_frame().rename(columns=
    {0:"missing values"})
```

#### Cabin Feature

```python
train = train.drop(["Cabin"],axis=1)
test = test.drop(["Cabin"],axis=1)
```

#### Ticket Feature

```python
train = train.drop(["Ticket"],axis=1)
test = test.drop(["Ticket"],axis=1)
```

#### Name Feature

```python
train = train.drop(["Name"],axis=1)
test = test.drop(["Name"],axis=1)
```

#### Embarked Features

```python
train = train.fillna({"Embarked" : "S"})

embarked_mapping = {"S":1, "C":2, "Q":3}
train["Embarked"] = train["Embarked"].map(embarked_mapping)
test["Embarked"] = test["Embarked"].map(embarked_mapping)
```

#### Fare Feature

```python
for x in range(len(test["Fare"])):
    if pd.isnull(test["Fare"][x]):
        pclass = test["Pclass"][x]
        test["Fare"][x] = round(train[train["Pclass"]==pclass]["Fare"].mean(),4)
    
train["FareBand"] = pd.qcut(train["Fare"],4, labels=[1,2,3,4])
test["FareBand"] = pd.qcut(test["Fare"],4,labels=[1,2,3,4])

train = train.drop(["Fare"],axis=1)
test = test.drop(["Fare"],axis=1)
```

#### Sex Feature

```python
sex_mapping = {"male": 0, "female":1}
train["Sex"] = train["Sex"].map(sex_mapping)
test["Sex"] = test["Sex"].map(sex_mapping)
```

#### Age Feature

```python
age_mapping = {'Infants': 0, 'Baby': 1, 'Child': 2, 'Teenager': 3,
               'Student': 4, 'Young Adult': 5, 'Adult': 6, 'Senior': 7}
train["AgeGroup"] = train["AgeGroup"].map(age_mapping)
test["AgeGroup"] = test["AgeGroup"].map(age_mapping)
```

# Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

predictors = train.drop(["Survived","PassengerId"],axis=1)
target = train["Survived"]
```

```python
x_train, x_val, y_train, y_val = train_test_split(predictors, target,
                    test_size=0.2, random_state=0)
```

### 1. Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report

gaussian = GaussianNB()
gaussian.fit(x_train,y_train)
y_pred = gaussian.predict(x_val)
acc_gaussian = round(accuracy_score(y_pred, y_val) * 100, 2)

print("Accuracy for Naive Bayes: ", acc_gaussian)
print("...........................On train Data...........................")
print(classification_report(y_train, gaussian.predict(x_train)))
print("...........................On test Data............................")
print(classification_report(y_val, gaussian.predict(x_val)))
print("...........................Confusion Matrix........................")
print(confusion_matrix(y_val, gaussian.predict(x_val)))
```

### 2. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_val)
acc_logreg = round(accuracy_score(y_pred, y_val) * 100, 2)
print("Accuracy for Logistic Regression : ", acc_logreg)
print("...........................On train Data...........................")
print(classification_report(y_train, logreg.predict(x_train)))
print("...........................On test Data............................")
print(classification_report(y_val, logreg.predict(x_val)))
print("...........................Confusion Matrix........................")
print(confusion_matrix(y_val, logreg.predict(x_val)))
```
